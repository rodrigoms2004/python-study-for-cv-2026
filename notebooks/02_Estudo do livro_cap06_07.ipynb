Factorial

````python
def fact_iter(n):
    """Assumes n an int > 0
    returns n!"""
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

def fact_rec(n):
    """Assumes n an int > 0
    returns n!"""
    if n == 1:
        return n
    else:
        return n * fact_rec(n - 1)


print(fact_iter(10))
print(fact_rec(10))
````

#### Finger exercise


````python
def harmonic_sum_rec(n):
    """Assumes n an int returns n = 1 + 1/2 + ... + 1/n"""
    if n == 1:
        return n
    else:
        return 1/n + harmonic_sum_rec(n - 1)
    
# An Harmonic SUM using numpy
def harmonic_sum_numpy(n):
    return float (np.sum(1 / np.arange(1, n + 1))) # by default it returns type numpy.float64


for i in range (1, 11):
    h_rec = harmonic_sum_rec(i)
    h_npy =  harmonic_sum_numpy(i)
    print(f'For n = {i}. Recursive {h_rec:.4f}, NumPy {h_npy:.4f}. Comparing: {h_rec == h_npy}')
````

### Fibonacci Numbers



````python
def fib(n):
    """Assumes n int >= 0
    Returns Fibonacci of n"""
    if n == 0 or n == 1:
        return 1
    else:
        return fib(n - 1) + fib(n - 2)
    
def test_fib(n):
    for i in range(n + 1):
        print(f'fib of {i} = {fib(i)}')

test_fib(5)
````

#### Finger exercise

When the implementation of fib is
used to compute fib(5), how many times does it compute the value
of fib(2) on the way to computing fib(5)?

````python
def fib(n):
    """Assumes n int >= 0
    Returns Fibonacci of n"""
    if n == 0 or n == 1:
        print(f'for n = {n} then n = {n}')
        return 1
    else:
        print(f'for n = {n} then fib(n - 1) = fib({n - 1}), fib(n - 2) = fib({n - 2})')
        return fib(n - 1) + fib(n - 2)
````

### Palindromes

````python
def is_palindrome(s):
    """Assumes s is a str
        Returns True if letters in s form a palindrome; 
        False otherwise. Non-letters and capitalization are ignored
    """

    def to_chars(s):
        s = s.lower()
        letters = ''
        for c in s:
            if c in 'abcdefghijklmnopqrstuvwxyz':
                letters = letters + c
        return letters
    
    def is_pal(s):
        if len(s) <= 1:
            return True
        else: 
            return s[0] == s[-1] and is_pal(s[1:-1])
        
    return is_pal(to_chars(s))
````

````python
palindrome_list = ['Anna', 'Bob', 'Civic', 'Dad', 'Deified', 'Detartrated', 'Eve', 'Kayak', 'Level']
non_palindrome_list = ['Alpha', 'energy', 'hannah']

for word in palindrome_list:
    print(is_palindrome(word))


for word in non_palindrome_list:
    print(is_palindrome(word))
````

#### Divide and conquer

````python
def is_palindrome(s):
    """Assumes s is a str
        Returns True if letters in s form a palindrome; 
        False otherwise. Non-letters and capitalization are ignored
    """

    def to_chars(s):
        s = s.lower()
        letters = ''
        for c in s:
            if c in 'abcdefghijklmnopqrstuvwxyz':
                letters = letters + c
        return letters
    
    def is_pal(s):
        print('     is_pal called with', s)
        if len(s) <= 1:
            print('     About to return True from base case')
            return True
        else: 
            answer = s[0] == s[-1] and is_pal(s[1:-1])
            print('     About to return', answer, 'for', s)
            return answer
        
    return is_pal(to_chars(s))
````

````python
print('Try dogGod')
print(is_palindrome('dogGod'))
print('Try doGood')
print(is_palindrome('doGood'))
````

### Global Variables

````python
def fib(x):
    """
    Assumes x an int >= 0
    Returns Fibonacci of x
    """
    global num_fib_calls
    num_fib_calls += 1
    if x == 0 or x == 1:
        return 1
    else:
        return fib(x - 1) + fib(x - 2)
    
def test_fib(n):
    for i in range(n+1):
        global num_fib_calls
        num_fib_calls = 0
        print(f'fib of {i} = {fib(i)}')
        print(f'fib called {num_fib_calls} times')


test_fib(6)
````
