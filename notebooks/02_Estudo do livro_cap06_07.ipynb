Factorial

````python
def fact_iter(n):
    """Assumes n an int > 0
    returns n!"""
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

def fact_rec(n):
    """Assumes n an int > 0
    returns n!"""
    if n == 1:
        return n
    else:
        return n * fact_rec(n - 1)


print(fact_iter(10))
print(fact_rec(10))
````

````python
def harmonic_sum_rec(n):
    """Assumes n an int returns n = 1 + 1/2 + ... + 1/n"""
    if n == 1:
        return n
    else:
        return 1/n + harmonic_sum_rec(n - 1)
    
# An Harmonic SUM using numpy
def harmonic_sum_numpy(n):
    return float (np.sum(1 / np.arange(1, n + 1))) # by default it returns type numpy.float64


for i in range (1, 11):
    h_rec = harmonic_sum_rec(i)
    h_npy =  harmonic_sum_numpy(i)
    print(f'For n = {i}. Recursive {h_rec:.4f}, NumPy {h_npy:.4f}. Comparing: {h_rec == h_npy}')
````

### Fibonacci Numbers



````python
def fib(n):
    """Assumes n int >= 0
    Returns Fibonacci of n"""
    if n == 0 or n == 1:
        return 1
    else:
        return fib(n - 1) + fib(n - 2)
    
def test_fib(n):
    for i in range(n + 1):
        print(f'fib of {i} = {fib(i)}')

test_fib(5)
````

### Palindromes

````python
def is_palindrome(s):
    """Assumes s is a str
        Returns True if letters in s form a palindrome; 
        False otherwise. Non-letters and capitalization are ignored
    """

    def to_chars(s):
        s = s.lower()
        letters = ''
        for c in s:
            if c in 'abcdefghijklmnopqrstuvwxyz':
                letters = letters + c
        return letters
    
    def is_pal(s):
        if len(s) <= 1:
            return True
        else: 
            return s[0] == s[-1] and is_pal(s[1:-1])
        
    return is_pal(to_chars(s))
````

````python
palindrome_list = ['Anna', 'Bob', 'Civic', 'Dad', 'Deified', 'Detartrated', 'Eve', 'Kayak', 'Level']
non_palindrome_list = ['Alpha', 'energy', 'hannah']

for word in palindrome_list:
    print(is_palindrome(word))


for word in non_palindrome_list:
    print(is_palindrome(word))
````

#### Divide and conquer

````python
def is_palindrome(s):
    """Assumes s is a str
        Returns True if letters in s form a palindrome; 
        False otherwise. Non-letters and capitalization are ignored
    """

    def to_chars(s):
        s = s.lower()
        letters = ''
        for c in s:
            if c in 'abcdefghijklmnopqrstuvwxyz':
                letters = letters + c
        return letters
    
    def is_pal(s):
        print('     is_pal called with', s)
        if len(s) <= 1:
            print('     About to return True from base case')
            return True
        else: 
            answer = s[0] == s[-1] and is_pal(s[1:-1])
            print('     About to return', answer, 'for', s)
            return answer
        
    return is_pal(to_chars(s))
````

### Global Variables

````python
def fib(x):
    """
    Assumes x an int >= 0
    Returns Fibonacci of x
    """
    global num_fib_calls
    num_fib_calls += 1
    if x == 0 or x == 1:
        return 1
    else:
        return fib(x - 1) + fib(x - 2)
    
def test_fib(n):
    for i in range(n+1):
        global num_fib_calls
        num_fib_calls = 0
        print(f'fib of {i} = {fib(i)}')
        print(f'fib called {num_fib_calls} times')


test_fib(6)
````

#### Finger exercise: 
Write a function that meets the specification

American Thanksgiving is celebrated on the fourth Thursday in November. In 2026, the holiday falls on November 26

Christmas is celebrated annually on December 25th

````python
# a GO HORSE code

def shopping_days(year):
    """year a number >= 1941 returns the number of days between U S Thanksgiving 
        and Christmas in year"""
    if year >= 1941:
        lastDayNovember = max(cal.monthcalendar(year, 11)[4::][0])
        return (lastDayNovember - find_thanksgiving(year)) + 25 # last day of november - thansgiven + christmas
    
print(shopping_days(2025))
````

Write a function that takes a year (>1957) as a parameter, and returns the number of days between
canadian Thanksgiving and christmas.

````python
def find_canadian_thanksgiving(year):
    month = cal.monthcalendar(year, 10)
    if month[0][cal.MONDAY] != 0:
        thanksgiving = month[1][cal.MONDAY]
    else:
        thanksgiving = month[2][cal.MONDAY]
    return thanksgiving

# print(find_canadian_thanksgiving(2025))


def canadian_shopping_days(year):
    """year a number >= 1958 returns the number of days between Canadian Thanksgiving 
        and Christmas in year"""
    if year >= 1958:
        lastDayOctober = max(cal.monthcalendar(year, 10)[4::][0])
        return (lastDayOctober - find_canadian_thanksgiving(year)) + 30 + 25 # last day of october - thansgiven + november + christmas
    

print(canadian_shopping_days(2025))
````
