#### Finger exercise: 
Write a program that prints the sum of the prime
numbers greater than 2 and less than 1000. Hint: you probably want
to use a for loop that is a primality test nested inside a for loop that
iterates over the odd integers between 3 and 999

```python
def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    
    index = 3
    while index * index <= n:
        if n % index == 0:
            return False
        index += 2

    return True
```

```python
total = 0
for c in range(2, 1000):
    if (is_prime(c)):
        total += c

print(total)
```

#### Approximate Solutions and Bisection Search

```python
# Approximation of the square root of x
x = 123456 # 25 and 0.25
epsilon = 0.01
step = epsilon**2
num_guesses = 0
ans = 0.0

while abs(ans**2 - x) >= epsilon and ans*ans <= x:
    ans += step
    num_guesses += 1

print(f'number of guesses = {num_guesses}')

if abs(ans**2 - x) >= epsilon:
    print(f'Failed on square root of {x}')
else:
    print(f'{ans} is close to square root of {x}')

```

```python
# Approximation of the square root of x
x = 123456 # 25 and 0.25
epsilon = 0.01
high = max(1, x)
num_guesses, low = 0, 0
ans = (high + low) / 2

while abs(ans**2 - x) >= epsilon:
    print(f'low={low}, high={high}, ans={ans}')
    num_guesses += 1
    
    if ans**2 < x:
        low = ans
    else:
        high = ans
    
    ans = (high + low) / 2
    
print(f'number of guesses = {num_guesses}')
print(f'{ans} is close to square root of {x}')
    
```

```python
# Using bisection search to estimate log base 2
# Find lower bound on ans
x = 10
lower_bound = 0
epsilon = 0.01
while 2**lower_bound < x:
    lower_bound +=1

low = lower_bound - 1
high = lower_bound + 1

# Perform bisection search
ans = (high + low) / 2
while abs(2**ans - x) >= epsilon:
    if 2**ans < x:
        low = ans
    else:
        high = ans
    
    ans = (high + low) / 2

print(f'{ans} is close to the log base 2 of {x}')
```

#### Finger exercise: 
The Empire State Building is 102 stories high. A
man wanted to know the highest floor from which he could drop an
egg without the egg breaking. He proposed to drop an egg from the
top floor. If it broke, he would go down a floor, and try it again. He
would do this until the egg did not break. At worst, this method
requires 102 eggs. Implement a method that at worst uses seven
eggs.

```python
# I have 102 stories high, and 7 eggs or 7 guesses, a binary search 

low = 0
high = 102
eggs_used = 0
eggs_available = 7
break_floor = 56

while low < high and eggs_used < eggs_available:
    mid = (low + high + 1 ) // 2 # rounded division
    eggs_used += 1

    if mid >= break_floor:
        high = mid - 1
    else:
        low = mid

    print(f'low={low}, mid={mid}, high={high}, eggs_used={eggs_used}')

print(f'Eggs used: {eggs_used} safe floor: {low}')

```

#### Newton–Raphson

```python
# Figure 3-7 Implementation of Newton–Raphson method
# Newton-Raphson for square root
# Find x such that x**2 - 24 is within epsilon of 0 , x**2 - k = 0
k = 24
epsilon = 0.01
guess = k / 2

while abs(guess**2 - k) >= epsilon:
    guess = guess - ( ( (guess**2) - k ) / (2*guess) )

print(f'Square root of {k} is about {guess}')
```
